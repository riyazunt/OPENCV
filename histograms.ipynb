```python
# allow to visualize pixal intensity distribution in an image

import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv.imread('Photos/cats.jpg')
cv.imshow('Cats', img)

gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
cv.imshow('Gray', gray)

grey_hist  = cv.calcHist([gray], [0], None, [256], [0,256])
# list of images, no.of channels, mask, bins, range
# plt.hist(grey_hist)

plt.figure()
plt.title('Greyscale Histogram')
plt.xlabel('Bins')
plt.ylabel('no.of pixels')
plt.plot(grey_hist)
plt.xlim([0,256])
plt.show()

cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# can create a mask and generate histogram only on the mask
```

```python
img = cv.imread('Photos/cats 2.jpg')
cv.imshow('Cats_2', img)

blank = np.zeros(img.shape[:2], dtype = 'uint8')

gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
cv.imshow('Gray', gray)

circle = cv.circle(blank.copy(), (img.shape[1]//2, img.shape[0]//2), 100, 255, -1)
# cv.imshow('Circle', circle)

masked_img  = cv.bitwise_and(gray, gray, mask=circle)
cv.imshow('Masked image', masked_img)

grey_hist  = cv.calcHist([gray], [0], masked_img, [256], [0,256])
# list of images, no.of channels, mask, bins, range
# plt.hist(grey_hist)

plt.figure()
plt.title('Greyscale Histogram')
plt.xlabel('Bins')
plt.ylabel('no.of pixels')
plt.plot(grey_hist)
plt.xlim([0,256])
plt.show()

cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# color histogram

img = cv.imread('Photos/cats.jpg')
cv.imshow('Cats', img)

blank = np.zeros(img.shape[:2], dtype = 'uint8')

# gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
# cv.imshow('Gray', gray)

circle = cv.circle(blank.copy(), (img.shape[1]//2, img.shape[0]//2), 100, 255, -1)
# cv.imshow('Circle', circle)

masked_img  = cv.bitwise_and(img, img, mask=circle)
cv.imshow('Masked image', masked_img)

# grey_hist  = cv.calcHist([gray], [0], masked_img, [256], [0,256])
# list of images, no.of channels, mask, bins, range
# plt.hist(grey_hist)

# plt.figure()
# plt.title('Greyscale Histogram')
# plt.xlabel('Bins')
# plt.ylabel('no.of pixels')
# plt.plot(grey_hist)
# plt.xlim([0,256])
# plt.show()

plt.figure()
plt.title('BGR Histogram')
plt.xlabel('Bins')
plt.ylabel('no.of pixels')
colors = ('b', 'g', 'r')
for i, col in enumerate(colors):
    hist = cv.calcHist([img], [i], None, [256], [0,256])
    plt.plot(hist, color = col)
    plt.xlim([0,256])
    
plt.show()    

cv.waitKey(0)
cv.destroyAllWindows()
```
