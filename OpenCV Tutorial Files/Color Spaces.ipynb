```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv.imread('Photos/park.jpg') # 3 channel BGR image
# cv.imshow('Park', img)

grey = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
# cv.imshow('Grey', grey)

hsv = cv.cvtColor(img, cv.COLOR_BGR2HSV)
# cv.imshow('HSV', hsv)

lab = cv.cvtColor(img, cv.COLOR_BGR2LAB)
# cv.imshow('LAB', lab)

rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)
# cv.imshow('RGB', rgb)

# plt.imshow(rgb)
# plt.show()


all_img = [img, grey, hsv, lab, rgb]
labels = ['IMAGE', 'GRAY', 'HSV', 'LAB', 'RGB']

plt.figure(figsize=(12,8), dpi=250)
for i in range(5):
    plt.subplot(2, 3, i+1)
    all_img[i] = cv.cvtColor(all_img[i], cv.COLOR_BGR2RGB)
    plt.imshow(all_img[i])
    plt.title(labels[i])
    plt.xticks([]), plt.yticks([])
    
plt.show()



# cv.imshow('IMAGE',all_img[1])
# plt.show()

cv.waitKey(0)
```
