```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# morphological operations are some simple operations based on the image shape
# erotion, dialation, opening and closing
# performed these operations on binary images

# need original image, structuring element or kernal which decides the nature of the operation
```

```python
img = cv.imread('Photos/balls.jpg', cv.IMREAD_GRAYSCALE) # or 0 to read image in greyscale
_, mask = cv.threshold(img, 247, 255, cv.THRESH_BINARY_INV)


kernal = np.ones([5,5], np.uint8)

# eg:- dialation used to remove black dots in mask below
dilation = cv.dilate(mask, kernal, iterations = 2 ) #by default 1
# kernal is a shape which we want to apply on the image


erosion  = cv.erode(mask, kernal, iterations = 2)
# erodes away the boundaries of background object
# when applied, a pixel in orginal image considered as one if all the pixel under kernal is one else it will be eroded (0)


opening = cv.morphologyEx(mask, cv.MORPH_OPEN, kernal, iterations = 2)
# actually erosion followed by dialation.

closing = cv.morphologyEx(mask, cv.MORPH_CLOSE, kernal, iterations = 2)
# actually dialation followed by erosion


titles = ['Image', 'mask', 'dilation', 'erosion', 'opening', 'closing']
images = [img, mask, dilation, erosion, opening, closing]


plt.figure(figsize=(12,8), dpi=250)
for i in range(6):
    plt.subplot(2, 3, i+1)
    plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])
    
plt.show()    
```

```python
img = cv.imread('Photos/balls.jpg', cv.IMREAD_GRAYSCALE) # or 0 to read image in greyscale
_, mask = cv.threshold(img, 247, 255, cv.THRESH_BINARY_INV)


kernal = np.ones([5,5], np.uint8)

# eg:- dialation used to remove black dots in mask below
dilation = cv.dilate(mask, kernal, iterations = 2 ) #by default 1
# kernal is a shape which we want to apply on the image


erosion  = cv.erode(mask, kernal, iterations = 2)
# erodes away the boundaries of background object
# when applied, a pixel in orginal image considered as one if all the pixel under kernal is one else it will be eroded (0)


opening = cv.morphologyEx(mask, cv.MORPH_OPEN, kernal, iterations = 2)
# actually erosion followed by dialation

closing = cv.morphologyEx(mask, cv.MORPH_CLOSE, kernal, iterations = 2)
# actually dialation followed by erosion

morph_grad = cv.morphologyEx(mask, cv.MORPH_GRADIENT, kernal, iterations = 2)
# difference b/w dialation and erosion of and image

top_hat = cv.morphologyEx(mask, cv.MORPH_TOPHAT, kernal, iterations = 2)
# actually differnce b/w input image and opening

black_hat = cv.morphologyEx(mask, cv.MORPH_BLACKHAT, kernal, iterations = 2)
# It is the difference between the closing of the input image and input image.


titles = ['Image', 'mask', 'dilation', 'erosion', 'opening', 'closing', 'morph_grad', 'top_hat', 'black_hat']
images = [img, mask, dilation, erosion, opening, closing, morph_grad, top_hat, black_hat]


plt.figure(figsize=(12,8), dpi=250)
for i in range(9):
    plt.subplot(3, 3, i+1)
    plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])
    
plt.show()    
```
