```python
capture = cv.VideoCapture(0)
fourcc = cv.VideoWriter_fourcc(*'XVID') # codec
out = cv.VideoWriter('sample_out.mp4', fourcc, 20.0,(640,480)) # codec, framespersecond and resolution

while (capture.isOpened()):
    isTrue, frame = capture.read() # ret
    
    if isTrue: # (ret == True)
        
        out.write(frame)
        cv.imshow('output', frame)   
    
        if cv.waitKey(25) & 0xFF == ord('q'):
            break
    else:
        break
        
capture.release() 
cv.destroyAllWindows()
cv.waitKey(60)
```
